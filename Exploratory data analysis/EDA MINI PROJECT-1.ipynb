##### Ques - 3
Converting the dtypes of desired columns for Data Analysis

All the variable except Value,Wage ,Contract Valid Until ,Height ,Weight , Joined and Release Clause are not in desired datatypes

lets change these variable to desired datatype.

```python
df['Value'].tail(200)
```

```python
def str_to_num(column):
    l = []
    for i in column:
        if len(i) == 2:
            l.append(int(i[-1]))
        else:
            num=i[1:-1]
            if i[-1] == 'M':
                l.append(float(num)*1000000)
            elif i[-1] == 'K':
                l.append(float(num)*1000)
            else:
                l.append(float(num)) 
    return  l
```

```python
df['Value']=str_to_num(df['Value'])
df['Wage'] = str_to_num(df['Wage'])
```

```python
df['Release Clause'].dtypes
```

```python
l=[]
for i in df['Release Clause']:
    if type(i)== str:
        num = i[1:-1]
        if i[-1]=='M':
            l.append(float(num)*1000000)
        elif i[-1]=='K':
            l.append(float(num)*1000)
        else:
            l.append(float(num))
    else:
        l.append(i)
```

```python
df['Release Clause'] = l
```

```python
df['Release Clause'].tail()
```

```python
df['Joined'].isnull().sum()
```

```python
df['Joined'].dtypes
```

```python
j = []
for i in df['Joined']:
    if type(i)== str:
        j.append(int((i.split(',')[-1])))
    else:
        j.append(i)
        
df['Joined'] = j
```

```python
df['Joined']
```

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
```

```python
df['Contract Valid Until']
```

```python
l = []
for i in df['Height']:
    if type(i)==str:
        ft = i.split("'")[0]
        inch = ''.join(i.split("'")[1:])
        inch = float(inch)*(1/12)
        num = float(ft) + inch
        l.append(num)
    else:
        l.append(i)
df['Height'] = l     
```

```python
df['Height']
```

```python
l = []
for i in df['Weight']:
    if type(i)==str:
        num = i[:-3]
        l.append(float(num))
    else:
        l.append(i)
df['Weight'] = l     
```

```python
df['Weight'].isnull().sum()
```
